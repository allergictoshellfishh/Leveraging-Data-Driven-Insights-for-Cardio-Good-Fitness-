# Data Preparation <a id = "data"></a>
The dataset that have been given is in.csv format. There are a few areas that has to be transformed for the data to be accessible for analysis. Here are some observations and changes made:

<u>Data Structure:</u> This file initially contained 184 rows with 10 columns. After I have filtered the data, data file now contains 180 rows with 10 columns. I have done exploration on possible data issues and rectified them where appropriate.

1. <u>Column Names:</u> Changed all the column names to standardised lower case and snakecase convention

2. <u>Rows with All Missing (NA) Values:</u> I removed the rows where all their values except for 'Product' and 'Branch', were missing (na). These rows were excluded to ensure data integrity and maximize the relevance of insights derived from the visualisations.

3. <u>Handle Missing (NA) Values:</u> I standardized all occurences of the string 'na' by converting them to NaN, ensuring uniform representation of missing values across the dataset.

4. <u>Edit Date Types:</u> I adjusted the data types of numerical columns to integers, allowing for seamless integration in data visualisation and analysis.

5. <u>Edit 'maritalstatus' Column:</u> In the original dataset, certain entries in the 'maritalstatus' column were 'S' and 'P' instead of 'Single' and 'Partnered'. To ensure consistency, I standardized all values to their full forms.

6. <u>Duplicated Rows:</u> I checked for duplicate rows and removed any that were found to maintain data integrity.

## Data Cleaning <a id = "cleaning"></a>

```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.gridspec import GridSpec
```

```python
# Importing Data
df = pd.read_csv('cardio_good_fitness.csv')

# Backup copy of dataframe
df_bkup = df.copy()

# Display basic info before cleaning
print("Before Cleaning:")
print(df.info())
print(df.head())
print("Shape before cleaning:", df.shape)
```

```python
# Step 1: Convert all columns naming to lowercase 
df.columns = df.columns.str.strip()
df.columns = [x.lower().replace(' ','_') for x in df.columns]

# Display the new column names 
print("Cleaned Column Names: ", df.columns.tolist())
```

```python
# Step 2: Handle Rows with All Missing (NA) Values 
# Drop rows where all values in the selected columns are NA
columns_to_check = ["age", "gender", "education", "maritalstatus", "usage", "fitness", "income", "miles"]
rows_to_remove = df[df[columns_to_check].apply(lambda row: (row == 'na').all(), axis=1)]
df.drop(index = rows_to_remove.index, inplace = True)
df.reset_index(drop = True, inplace = True)

# Display result
print("Rows where all values except 'product' and 'branch' were NA have been removed.")
print(df.head())
print()
print("Shape after deleting rows will all null values:", df.shape)
```

```python
# Step 3: Handle Missing (NA) Values

# Determine which rows has NaN
df.isnull().any()
df.replace('na', np.nan, inplace = True)
print (df[['age', 'usage']].isna().sum())
print()

# Since the number of NaN is more than 5%, I will replace instead of delete NaN
df['age'] = pd.to_numeric(df['age'], errors = 'coerce')
df['usage'] = pd.to_numeric(df['usage'], errors = 'coerce')
df.update(df['age'].fillna(round(df['age'].mean())))
df.update(df['usage'].fillna(round(df['usage'].mean())))
print(df.head())
```

```python
# Step 4: Edit Date Types
df['age'] = df['age'].astype(int)
df['education'] = df['education'].astype(int)
df['usage'] = df['usage'].astype(int)
df['fitness'] = df['fitness'].astype(int)
df['income'] = df['income'].astype(int)
df['miles'] = df['miles'].astype(int)
print(df.info())
```

```python
# Step 5: Standardize Values in 'maritalstatus' Column
df['maritalstatus'] = df['maritalstatus'].replace('S', 'Single')
df['maritalstatus'] = df['maritalstatus'].replace('P', 'Partnered')
print(df['maritalstatus'].unique()) 
print(df.head())
```

```python
# Step 6: Determine if there are Any Duplicated Rows
num_duplicates = df.duplicated().sum()
print("Total Duplicate Rows:", num_duplicates)
if num_duplicates > 0:
    df.drop_duplicates(inplace = True)
    df.reset_index(drop = True, inplace = True)
else:
    print("There are no duplicates")
```

```python
# Step 7: Display Final Cleaned Data
print(df.info())
print(df)
```

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;i. Bullet chart that shows popularity of treadmill model among differnet income groups.

```python
# Visual b)i)

# Set style
sns.set_style("whitegrid")

# Define treadmill models
treadmill_models = df['product'].unique()

# Calculate summary statistics for income per treadmill model
summary_stats = df.groupby('product')['income'].agg(['min', 'mean', 'max'])

# Plot bullet chart
fig, ax = plt.subplots(figsize=(10, 6))
bar_color = 'skyblue'  
marker_color = 'red'   
for i, model in enumerate(treadmill_models):
    min_val = summary_stats.loc[model, 'min']
    mean_val = summary_stats.loc[model, 'mean']
    max_val = summary_stats.loc[model, 'max']
    ax.barh(i, max_val - min_val, left = min_val, color = bar_color, height = 0.4, alpha = 0.7)
    ax.plot(mean_val, i, '|', color=marker_color, markersize=15)
    ax.text(mean_val, i, f'   ${mean_val:,.0f}', va = 'center', ha = 'left', fontsize = 10, color = 'red')

# Customize plot
ax.set_yticks(range(len(treadmill_models)))
ax.set_yticklabels(treadmill_models)
ax.set_xlabel("Income Level ($)", fontsize = 12)
ax.set_title("Treadmill Model Popularity by Income Level", fontsize = 14)
plt.grid(axis = 'x', linestyle = '--', alpha = 0.7)

# Show plot
plt.show()
```

Each horizontal bar represents an income range for customers who purchased a specific treadmill model. The red line represents the mean income level for each treadmill model. Wider bars indicate a broader range of income levels, while narrower bars suggest a more specific customer segment. The bullet chart visualizes the income distribution of customers who purchased different treadmill models. The light blue bars represent the income range (min to max) of customers for each treadmill model, while the red markers indicate the average income of buyers for each model.


Overall, TM798 has the widest income range and the highest average income of 75,442 USD, suggesting it is the most popular treadmill among high-income customers. TM195 and TM498 have lower income ranges and lower average incomes of 46,418 and 48,974 USD respectively, indicating they appeal to mid-to-lower income groups. TM195 is purchased by customers with a low average income suggesting it is a budget-friendly option, while TM498 is purchased by customers with a slightly higher income range than TM195 of around 32,000 - 67,000 USD, with a slightly higher average income. This suggest that the treadmill might offer more features while still remaining affordable. Lastly, TM798 has the highest price point, appealing to high-income customers, with some buyers earning over 100,000 USD. This suggests it is a premium treadmill choice.


This graph answers the exploratory question bi - Which treadmill model is the most popular among different income groups?

```python
def visualci(ax):
    branch_sales = df['branch'].value_counts()
    colors = sns.color_palette("Set2", n_colors = len(branch_sales))
    ax.pie(branch_sales, labels = branch_sales.index, autopct = '%1.1f%%', colors = colors, startangle = 140)
    ax.set_title("Treadmill Sales by Branch", fontsize = 14)

def visualcii(ax):
    ax = sns.boxplot(data = df, x = 'branch', y = 'age', hue = 'branch', legend = False, palette = 'Set2', ax = ax)
    medians = df.groupby('branch')['age'].median()
    for i, median in enumerate(medians):
        ax.text(i, median + 1, f'{median:.1f}', ha = 'center', va = 'bottom', fontsize = 12, color = 'black')
    ax.set_title("Age Distribution by Branch", fontsize = 14)
    ax.set_xlabel("Branch", fontsize = 12)
    ax.set_ylabel("Age", fontsize = 12)

def visualciii(ax):
    income_summary = df.groupby('branch')['income'].agg(['min', 'mean', 'max'])
    income_summary.columns = ['Min Income', 'Avg Income', 'Max Income']
    sns.heatmap(income_summary, annot=True, fmt = ".0f", cmap = "Blues", linewidths = 0.5, ax = ax)
    ax.set_title("Income Levels Across Branches", fontsize = 14)
    ax.set_xlabel("Income Statistics", fontsize = 12)
    ax.set_ylabel("Branch", fontsize = 12)

def visualciv(ax):
    sns.swarmplot(data = df, x = 'branch', y = 'miles', hue = 'branch', legend = False, palette = 'Set2', ax = ax)
    ax.set_title("Expected Miles Per Week by Branch", fontsize = 14)
    ax.set_xlabel("Branch", fontsize = 12)
    ax.set_ylabel("Miles Per Week", fontsize = 12)

def visualcv(ax):
    marital_counts = df.groupby(['branch', 'maritalstatus']).size().unstack()
    marital_proportions = marital_counts.div(marital_counts.sum(axis = 1), axis = 0)
    ax = marital_proportions.plot(kind = 'bar', stacked = True, colormap = 'Set2', ax = ax)
    ax.set_title("Marital Status By Branch", fontsize = 14)
    ax.set_xlabel("Branch", fontsize = 12)
    ax.set_ylabel("Proportion of Customers", fontsize = 12)
    ax.legend(title="Marital Status", fontsize = 10)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

#Create a figure to be used as a dashboard
fig = plt.figure(figsize = (16, 10))
gs = GridSpec(nrows = 2, ncols = 3, figure = fig, hspace = 0.3)

# Define subplots for each visualization
ax1 = fig.add_subplot(gs[0, 0]) 
ax2 = fig.add_subplot(gs[0, 1])  
ax3 = fig.add_subplot(gs[0, 2])  
ax4 = fig.add_subplot(gs[1, :2])  
ax5 = fig.add_subplot(gs[1, 2]) 

# Call functions to plot each visual
visualci(ax3)
visualcii(ax2)
visualciii(ax5)
visualciv(ax4)
visualcv(ax1)

# Set title of dashboard
fig.suptitle("Regional Sales Insights", fontsize = 22)

# Display plot
plt.show()
```

This dashboard provides a comprehensive analysis of treadmill sales across different branches, customer demographics, and regional income levels. By visualizing sales distribution, age demographics, expected treadmill usage, and income levels, businesses can tailor their marketing and product strategies to different regions effectively.

The pie chart at the top right compares the overall treadmill sales across the branches. They can determine when a branch is falling behind or doing exceptionally well and have more marketing for those branches. Currently, the relatively even distribution suggests a balanced market demand across branches, but North has the lowest sales share, indicating potential for increased market penetration.  
The box plot with median value stated at the top center shows the age distribution of customers across the different branches. Overall, the median age of customers varies slightly across branches indicating that majority of their customers are about the same age. Additionally, the interquartile range is similar across branches, suggesting age distribution is relatively consistent across regions. They can hence target their marketing to customers of these age range.  
The stacked bar chart at the top left shows the proportion of single and partnered customers across the branches. Overall, the data is fairly consistent, with partnered customers forming a larger segment on all locations. Hence indicating more marketing towards partnered customers.  
The heatmap at the bottom right shows the income of customers across the different branches. It can be seen that the average income is all relatively similar, indicating a broad middle-class customer base. However there is differences in the minimum and maximum income indicating possible wealthier customers at certain branches.  
Lastly, the swarm plot at the bottom left shows the miles customers expect to run per week by branch. Most customers expect to run between 50 to 150 miles per week, but some extreme users exceed 300 miles per week. Based on the expected miles to run, they can target marketing of the different treadmill models based on their features.

As such, this dashboard can provide the marketing research team with multiple actionable insights. They can focus more of their marketing on the North branch due to them having the lowest sales. Additionally, by relating to the other graphs, they can know which are the most suitable features to promote at the respective stores. For the West branch they can target younger customers through tech-driven fitness marketing. For the East branch, they can target premium treadmill sales. For the low-mileage users from the south branch, they can promote casual fitness programs and entry-level treadmill models.

```python
def visualdi(ax):
    df['fitness'] = df['fitness'].astype(str)
    sns.histplot(data = df, x = 'usage', hue = 'fitness', bins = 10, kde = True, multiple = "stack", palette = "Set2", ax = ax)
    ax.set_title("Distribution of Weekly Treadmill Usage by Fitness Rating", fontsize = 14)
    ax.set_xlabel("Weekly Treadmill Usage (Hours)", fontsize = 12)
    ax.set_ylabel("Number of Customers", fontsize = 12)

def visualdiii(ax):
    sns.regplot(data = df, x = "age", y = "miles", scatter = True, line_kws = {"color": "red", "linewidth": 2}, ax = ax)
    ax.set_title("Scatter Plot of Weekly Miles Across Different Ages", fontsize = 14)
    ax.set_xlabel("Age", fontsize = 12)
    ax.set_ylabel("Average Weekly Miles", fontsize = 12)

def visualdiv(ax):
    fitness_order = [1, 2, 3, 4, 5]
    sns.boxplot(data = df, x = 'fitness', y = 'usage', hue = 'fitness', palette = 'Set2', order = fitness_order, ax = ax)
    ax.set_title("Treadmill Usage Distribution Across Different Fitness Levels", fontsize = 14)
    ax.set_xlabel("Self-Rated Fitness Level (1-5)", fontsize = 12)
    ax.set_ylabel("Weekly Treadmill Usage (Hours)", fontsize = 12)

fig = plt.figure(figsize = (16, 10))
gs = GridSpec(nrows = 2, ncols = 3, figure = fig, hspace = 0.3)

# Define subplots for each visualization
ax1 = fig.add_subplot(gs[0:, :2]) 
ax2 = fig.add_subplot(gs[0, 2])  
ax3 = fig.add_subplot(gs[1, 2])  


# Call functions to plot each visual
visualdi(ax1)
visualdiii(ax2)
visualdiv(ax3)


# Set title of dashboard
fig.suptitle("Fitness & Treadmill Usage Patterns", fontsize = 22)

# Display plot
plt.show()
```

This dashboard provides an in-depth view of how treadmill usage varies across different fitness levels, age groups, and education levels. It examines key behavioral patterns, helping businesses understand which customer segments use treadmills the most and how treadmill usage correlates with fitness, age, and education.

The stacked histogram with the kernel density estimate at the left shows the weekly treadmill usage by different fitness ratings based on the number of customers. Overall, most customers use treadmills for 3 to 4 hours per week, regardless of fitness level.Those with a higher fitness rating  mainly cluster at 3 to 4 hours a week, with some customers using the treadmill for longer hours while customers with lower fitness ratings tend to have a broad distribution, showing that those with lower fitness ratings are either doing it casually or doing it to get fit.  
The scatter plot at the top right with the best fit line shows the weekly miles ran by customers of different ages. Overall, younger and older customers both show a wide range of weekly miles, but there is no strong correlation between age and treadmill mileage. Most users cluster around 50 to 150 miles per week, with a few outliers running over 300 miles weekly. The trendline shows a slight increase in mileage as age increases, suggesting that some older users may maintain a dedicated treadmill routine.  
The box plot at the bottom right shows the weekly treadmill usage based on customers self rated fitness level. It can be seen that as the fitness ratings increases, the treadmill usage increases from around 2-3 hours per week for the lower fitness levels to those with higher fitness levels consistently using the treadmill for more than 5 hours a week. Users with a fitness rating of 5 show the widest range of treadmill usage, indicating that even the fittest customers have varied treadmill habits. This can be compared to the stacked histogram to see the total count of customers to know to target customers of which specific fitness level. 

As such, this dashboard can provide the marketing research team with multiple actionable insights. For users of lower fitness, they can be encouraged to increase their treadmill usage through loyalty rewards, guided classes and interactive challengers. For users of higher fitness, they can offer them premium treadmills with advanced workout features and performance tracking tools. Since the mileage is not age-dependent, they can focus on overall user goals such as weight loss, endurance and more. The Fitness & Treadmill Usage Patterns Dashboard reveals key trends in treadmill engagement across fitness levels, age groups, and education levels.
